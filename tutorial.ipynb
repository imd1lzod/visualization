```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
# plt.plot(x,y, label="xxx", color="yellow", linewidth=5, marker=".", markersize=25, markeredgecolor='green', linestyle='--')

# Shorthand notation -->
plt.plot(x,y, 'y.--', label="old")
plt.plot(x,[2,4,6,3,8], 'r.-', label="new")

plt.title("First graph")
plt.xlabel("x")
plt.ylabel("y")
# plt.xticks([1,5,10])
# plt.yticks([1,2,3])
plt.legend()
plt.grid(True)
fig, ax = plt.subplots(2,2)
ax[0,0].plot([1,2,3], [5,6,7])

plt.scatter([1,2], [12,44])

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,5,4,5]

plt.scatter(x, y)
plt.show()
```

```python
# Bar Chart
names = ['Sport', 'Education', 'Tourism']
values = [10, 20, 30]
plt.bar(names, values)
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.show()
```

```python
# Pie Chart
sizes = [10, 35, 30, 25]
plt.pie(sizes, labels=['Math', 'English', 'History', 'Biology'])
plt.show()
```
